```python
# Step 1: Calculate mean temperature across all times
mean_temp = ds1.skt.mean(dim='time')

# Step 2: Convert temperature from Kelvin to Celsius (optional but recommended)
mean_temp_celsius = mean_temp - 273.15

# Step 3: Use geopandas to read the district boundaries
import geopandas as gpd
districts = gpd.read_file('D:/FYP/active/shape/District_Boundary.shp')

# Step 4: Convert xarray to a pandas DataFrame with coordinates
import numpy as np
lons, lats = np.meshgrid(mean_temp.longitude, mean_temp.latitude)
df_temp = pd.DataFrame({
    'longitude': lons.flatten(),
    'latitude': lats.flatten(),
    'temperature': mean_temp_celsius.values.flatten()
})

# Step 5: Convert to GeoDataFrame
from shapely.geometry import Point
gdf_temp = gpd.GeoDataFrame(
    df_temp, 
    geometry=[Point(xy) for xy in zip(df_temp.longitude, df_temp.latitude)],
    crs="EPSG:4326"
)

# Step 6: Spatial join and calculate mean temperature per district
district_temps = gpd.sjoin(districts, gdf_temp, how='left', predicate='contains')
district_avg = district_temps.groupby(['DISTRICT'])['temperature'].mean().reset_index()

# Step 7: Merge back with original districts GeoDataFrame
districts_with_temp = districts.merge(district_avg, on='DISTRICT')

# Save to file if needed
districts_with_temp.to_file('district_avg_temperatures.shp')
```

```python
# Calculate mean temperature across time dimension
mean_temp = ds1.skt.mean(dim='time')

# Convert from Kelvin to Celsius
mean_temp_celsius = mean_temp - 273.15

# Create point-based temperature data
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Create points from temperature grid
lons, lats = np.meshgrid(mean_temp.longitude, mean_temp.latitude)
temp_df = pd.DataFrame({
    'longitude': lons.flatten(),
    'latitude': lats.flatten(),
    'temperature': mean_temp_celsius.values.flatten()
})

# Convert to GeoDataFrame
temp_gdf = gpd.GeoDataFrame(
    temp_df, 
    geometry=[Point(xy) for xy in zip(temp_df.longitude, temp_df.latitude)],
    crs="EPSG:4326"
)

# Read districts shapefile
districts = gpd.read_file('D:/FYP/active/shape/District_Boundary.shp')

# Spatial join and calculate district means
district_temps = gpd.sjoin(districts, temp_gdf, how='left', predicate='contains')
result = district_temps.groupby('DISTRICT')['temperature'].mean().round(2).reset_index()

# Sort by temperature and display nicely
result = result.sort_values('temperature', ascending=False)
print("\nDistrict Average Temperatures (°C):")
print("==================================")
for _, row in result.iterrows():
    print(f"{row['DISTRICT']:<20} {row['temperature']:>6.2f}°C")
```

```python
# Get coordinate arrays
lons = ds1.longitude.values
lats = ds1.latitude.values

# Print dimensions
print(f"Number of longitudes: {len(lons)} ({lons[0]} to {lons[-1]})")
print(f"Number of latitudes: {len(lats)} ({lats[0]} to {lats[-1]})")

# Create meshgrid
lons_grid, lats_grid = np.meshgrid(lons, lats)

# Calculate total points
total_points = len(lons) * len(lats)
flattened_points = lons_grid.size

print(f"\nGrid dimensions: {lons_grid.shape}")
print(f"Total grid points: {total_points}")
print(f"Flattened array size: {flattened_points}")

# Optional: Quick visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(lons_grid.flatten(), lats_grid.flatten(), 
           alpha=0.5, s=1, c='blue')
plt.title(f'Distribution of {total_points} Grid Points')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
